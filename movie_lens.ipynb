```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
movies_path = 'movielens/movies.dat'
data_movies = pd.read_csv(movies_path, header=None, names=['movie_id', 'title', 'genres'], sep='::')
data_movies
```

```python
users_path = 'movielens/users.dat'
data_users = pd.read_csv(users_path, header=None, names=['user_id', 'gender', 'age', 'occupation', 'zipcode'], sep='::')
data_users
```

```python
ratings_path = 'movielens/ratings.dat'
data_ratings = pd.read_csv(ratings_path, header=None, names=['user_id', 'movie_id', 'ratings', 'timestamp'], sep='::')
data_ratings
```

```python
data_movies['genres'].str.split('\\|')
```

```python
data_users['gender'].unique()
```

```python
data = pd.merge(pd.merge(data_ratings, data_users), data_movies)
data
```

```python
data['title'].value_counts()
```

```python
# popular movies
# movies that have got more than 500 people who have rated for it
s1 = data['title'].value_counts()
s1
```

```python
s2 = s1[s1 > 500]
s2
```

```python
popular_film_titles = s2.index
popular_film_titles
```

```python
data
```

```python
# get the top 10 films that were loved by the females
d1 = data[['title', 'gender', 'ratings']]
d1
```

```python
d2 = d1.pivot_table('ratings', aggfunc='mean', index='title', columns=['gender'])
d2
```

```python
d3 = d2.loc[popular_film_titles]
d3
```

```python
d4 = d3.sort_values(by='F', ascending=False)
d5 = d4.head(10)
d5
```

```python
d5.plot(kind='bar')
plt.show()
```

```python
# get the top 10 films that were loved by the males
d6 = d3.sort_values(by='M', ascending=False)
d6
```

```python
# get the top 10 films where there was max diagreement between males and females
d3
```

```python
d3['diff'] = d3['F'] - d3['M']
d3['diff'] = d3['diff'].abs()
d3
```

```python
d7 = d3.sort_values(by='diff', ascending=True)
d7
```
