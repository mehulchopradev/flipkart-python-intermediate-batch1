```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
path = 'site-logs/session-logs.csv'
data = pd.read_csv(path, header=None, names=['username', 'date', 'time', 'action'])
data
```

```python
data['username'].value_counts()
```

```python
data['action'].unique()
```

# Number of times the different users have logged out from the site

```python
d1 = data[['username', 'action']]
d1
```

```python
d2 = d1[d1['action'] == 'out']
d2
```

```python
d3 = d2['username']
d3
```

```python
d4 = d3.value_counts()
d4
```

```python
d4.plot(kind='barh')
plt.show()
```

```python
d2
```

```python
d5 = d2.groupby('username')
d5
```

```python
d6 = d5.count()
d6
```

```python
d7 = d6.sort_values(by='action', ascending=False)
d7
```

# Number of times the different users have logged in and logged out

```python
d1
```

```python
d8 = d1.copy()
d8
```

```python
d8['total'] = 1
d8
```

```python
d9 = d8.groupby(['username', 'action'])
d9
```

```python
d10 = d9.sum()
d10
```

```python
d11 = d10.unstack()
d11
```

```python
d11.plot(kind='bar', stacked=True)
# plt.show()
plt.tight_layout()
plt.savefig('site-sessions.pdf', format='pdf')
```
